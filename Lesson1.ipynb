задача:
Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1..

```python
# 1. импортируем библиотеки

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# зададим массив значений переменной х
x=np.linspace(-4*np.pi, 4*np.pi, 180)

# определим функцию
def f(k,x):
    return np.cos(k*x)
```

```python
# Зададим коэффициенты:
k1=1
k2=k1 * float(input('Введите кратность коэффициента "k2"'))

# рисуем график:
plt.xlabel("Ось Х")
plt.ylabel("Ось У")
plt.grid(True)
plt.plot(x, f(k1,x), 'ro', x, f(k2,x), 'b--')
#plt.show
```

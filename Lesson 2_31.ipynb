```python
#Нарисуйте график функции:   
#для некоторых (2-3 различных) значений параметров k, a, b
 
# импортируем библиотеки
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# зададим массив значений переменной х
x=np.linspace(-2*np.pi, 2*np.pi, 180)

# определим функцию
def f(k,a,b):
    return k*np.cos(x-a)+b
```

```python
# Зададим коэффициенты:
k1 = 1
k2 = 2
k3 = 3
a1 = 7
a2 = 8
a3 = 9
b1 = 4
b2 = 5
b3 = 6


# рисуем график:
plt.xlabel("Ось Х")
plt.ylabel("Ось У")
plt.grid(True)
plt.plot(x, f(k1,a1,b1), x, f(k2,a2,b2), x, f(k3,a3,b3))
#plt.show
```

# 5. Дополните код расчетом коэффициента корреляции x и y по формуле

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
r = 0.85
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n

xm = np.sum(x)/n
ym = np.sum(y)/n
r = np.sum((x-xm)*(y-ym))/np.sqrt(np.sum((x-xm)**2)*np.sum((y-ym)**2))

#A = np.vstack([x, np.ones(len(x))]).T
#a1, b1 = np.linalg.lstsq(A, y)[0]
#print(a, b)
#print(a1, b1)

print (f'коэффициент корреляции R= {r:.4f}')
mat_r = np.corrcoef(x, y)
print (f'\nматрица корреляции веторов x и y:')
print (f'{mat_r[0,0]:.4f}  {mat_r[0,1]:.4f}')
print (f'{mat_r[1,0]:.4f}  {mat_r[1,1]:.4f}')
plt.plot([0, 1], [b, a + b])
plt.show()
```

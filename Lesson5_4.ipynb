# 4. Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k


```python
import numpy as np
import itertools
```

```python
n = 0
for p in itertools.permutations("abcdefgij",3):
    n += 1
    print(n, ''.join(str(x) for x in p))   
```

```python
n = 0
for p in itertools.combinations("abcdefgij",3):
    n += 1
    print(n, ''.join(p))   
```

```python
n = 0
for p in itertools.product("abcdefgij",repeat=3):
    n += 1
    print(n, ''.join(p))   
```
